## Introduction

FreeCAD is a powerful, free, and open-source parametric 3D CAD (Computer-Aided Design) software that caters to a wide range of users, from hobbyists to professional engineers, architects, and designers. Unlike many commercial CAD tools, FreeCAD is highly flexible and modular, allowing users to adapt it to their specific needs through workbenches, scripts, and plugins. It is ideal for precision modeling tasks, enabling users to create complex designs for engineering, architecture, and product design.

One of FreeCAD's standout features is its parametric modeling capability. This means that every aspect of a design is defined by parameters, which can be adjusted to modify the model dynamically. Whether you're designing mechanical components, creating architectural blueprints, or prototyping for 3D printing, FreeCAD offers the tools you need to achieve your goals.

In this tutorial, you will learn how to install FreeCAD, navigate its interface, and create basic 3D models. As you progress, you'll discover the powerful features and customization options that make FreeCAD a compelling alternative to proprietary CAD software.

# Setup for python



```python
import sys
import os

sys.path.append('/usr/lib/freecad/lib')
```

```python


# Import FreeCAD modules
import FreeCAD
import FreeCADGui  # For GUI functionality (optional)
import Part  # For part modeling
import Draft  # For 2D drafting

# Example: Create a simple cube in FreeCAD
doc = FreeCAD.newDocument("TestDoc")
cube = doc.addObject("Part::Box", "Box")
cube.Length = 50
cube.Width = 10
cube.Height = 10
doc.recompute()

# Save the file
output_path = os.path.join(os.getcwd(), "test_cube.FCStd")
doc.saveAs(output_path)

print(f"Model saved to {output_path}")
```

```python
import Part
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Export as STL
stl_path = "test_cube.stl"
cube.Shape.exportStl(stl_path)
print(f"STL file saved to {stl_path}")

# Optional visualization (requires STL parser)
# Install stl library: pip install numpy-stl
from stl import mesh

# Load STL
cube_mesh = mesh.Mesh.from_file(stl_path)

# Plot with matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.add_collection3d(Poly3DCollection(cube_mesh.vectors, alpha=0.7))

# Set scale and show
scale = cube_mesh.points.flatten("C")
ax.auto_scale_xyz(scale, scale, scale)
plt.show()
```
